# 台灣教育資源分配不均之現況探討

藉由分析開放式資料，探討應屆畢業生對於升學的選擇，進而延伸到主題－－教育資源不均等，並試圖提出可行之解決方案。

## 研究動機

我們很好奇教學資源、城鄉差距對升學的影響，又台灣的公、私立大學不知凡幾，所以我們想針對各縣市別（設立別、性別）來探討其相關性。

然而教學資源、城鄉差距等等的牽涉到非常多因素，因此我們想藉由升學統計資料研究台灣現在的教學環境與趨勢。

## 各縣市應屆畢業生升學情形

將各縣市應屆畢業生人數分類，並按比例製成圓餅圖。圖上的綠色區塊代表就讀國立大學比例；紅色區塊代表就讀軍警校比例；藍色區塊代表就讀私立大學比例；黃色區塊代表赴海外求學的比例。可以從圓餅圖中，比較各縣市就讀不同種類的學校的傾向。

經過比較後發現：資源與社會環境是影響各縣市應屆畢業生選擇的主因。例如就讀軍警校比例最高的縣市偏向於南部地區，但國內共八所軍警相關大學，一半分布於高雄，另一半則分布在台北地區。然而就讀比例最高的卻是在南部縣市而不是北部縣市，因為北部教學資源豐富，相對來說有較多的選擇，比起軍警校，國立大學是相對比較好的選擇。

```python
#台北市，國立大學、軍警校、私立大學、海外
import matplotlib.pyplot as plt
sizes=[11047,121,16703,603]#各區塊之大小
colors=["#99CC99","#CC3333","#6699CC","#FFCC99"]#各區塊之顏色，用色碼顯示
plt.pie(sizes,labeldistance=1.1,autopct="%2.2f%%",startangle=90,pctdistance=0.6,colors=colors)#圓餅圖的元素
plt.axis("equal")#把圖形變成正圓
plt.show()
```

## 各縣市就讀台大人數的比例

北部教育資源相對豐富，我們上網找到這份資料，佐證我們的觀點。透過資料分析，發現就讀台大的直轄市應屆畢業生總和佔了80%，由此可見，直轄市之教育資源相對其他縣市豐富許多。然而台北地區應屆畢業生就讀台大比例超過50%，可以顯著看出，台北地區的教育資源是全台最豐富的，也顯現出台灣教育資源分配不均的情形，這也是我們認為目前教育情形的問題之一。

```python
import matplotlib.pyplot as plt
import pandas as pd
csvdata9=pd.read_csv("台大.csv")#讀取檔案

x=csvdata9['district']#使用資料中的地區作為x軸
y=csvdata9['2011-2014']#使用就讀比例作為y軸
plt.figure(figsize=(12,6))#設定圖表大小
plt.bar(x,y,label="student%",width=0.4,align = "edge",color="#804040")#柱狀圖的元素
plt.title("2011-2014 NTU student")#圖表標題
plt.xticks(rotation=90)
plt.legend()
```

## 102~107年不同性別的升學情形

我們還分析了102～107年性別與就讀國立、私立大學的人數趨勢。
從資料中可以發現，無論是高中、職、綜高，女性升學入私校之數量幾乎都大於男性，但就讀公立大學的數量則都略遜於男性，我們懷疑是早期重男輕女之性別偏見遺毒。雖然近年來教育普及程度已大幅提升，但從趨勢圖裡仍可以發現一點性別偏見的蛛絲馬跡。

另各縣市就讀私立大學之人數皆明顯大於國立，但近年來受少子化影響，就讀私校之人數明顯不如既往，而國立大學幾乎未受影響。此處可推斷：私校將面臨合併、減系、消失之危機。這也是台灣逐漸浮現的教育問題，因為少子化而間接使的大學數量減少，進一步的加劇了教育資源不均等的現象。

```python
#高中
import matplotlib.pyplot as plt

year=[102,103,104,105,106,107]#設定年份
public_male=[19425,20203,19547,19498,19563,18935]#男性就讀公立大學人數
public_female=[17875,18058,17591,17465,17822,17444]#女性就讀公立大學人數
private_male=[27813,27972,24879,23007,24192,23202]#男性就讀私立大學人數
private_female=[31836,31832,30123,28017,29944,29105]#女性就讀私立大學人數
plt.plot(year,public_male,label="public,male",color="#B766AD")#折線圖元素
plt.plot(year,public_female,label="public,female",color="#00AEAE")#折線圖元素
plt.plot(year,private_male,label="private,male",color="#003D79")#折線圖元素
plt.plot(year,private_female,label="private,female",color="#FFA042")#折線圖元素
plt.legend()
plt.xlim(101,108)#設定x軸區間
plt.ylim(7000,38000)#設定y軸區間
plt.title('High School')#設定圖表標題
plt.xlabel('Year')#設定x軸標籤
plt.ylabel('Student Number')#設定y軸標籤
plt.show()
```
