```python
#高中
import matplotlib.pyplot as plt

year=[102,103,104,105,106,107]
public_male=[19425,20203,19547,19498,19563,18935]
public_female=[17875,18058,17591,17465,17822,17444]
private_male=[27813,27972,24879,23007,24192,23202]
private_female=[31836,31832,30123,28017,29944,29105]
plt.plot(year,public_male,label="public,male")
plt.plot(year,public_female,label="public,female")
plt.plot(year,private_male,label="private,male")
plt.plot(year,private_female,label="private,female")
plt.legend()
plt.xlim(101,108)
plt.ylim(7000,38000)
plt.title('High School')
plt.xlabel('Year')
plt.ylabel('Student Number')
plt.show()
```
