```python
#高中
import matplotlib.pyplot as plt

year=[102,103,104,105,106,107]#設定年份
public_male=[19425,20203,19547,19498,19563,18935]#男性就讀公立大學人數
public_female=[17875,18058,17591,17465,17822,17444]#女性就讀公立大學人數
private_male=[27813,27972,24879,23007,24192,23202]#男性就讀私立大學人數
private_female=[31836,31832,30123,28017,29944,29105]#女性就讀私立大學人數
plt.plot(year,public_male,label="public,male",color="#B766AD")#折線圖元素
plt.plot(year,public_female,label="public,female",color="#00AEAE")#折線圖元素
plt.plot(year,private_male,label="private,male",color="#003D79")#折線圖元素
plt.plot(year,private_female,label="private,female",color="#FFA042")#折線圖元素
plt.legend()
plt.xlim(101,108)#設定x軸區間
plt.ylim(7000,38000)#設定y軸區間
plt.title('High School')#設定圖表標題
plt.xlabel('Year')#設定x軸標籤
plt.ylabel('Student Number')#設定y軸標籤
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
csvdata9=pd.read_csv("台大.csv")#讀取檔案

x=csvdata9['district']#使用資料中的地區作為x軸
y=csvdata9['2011-2014']#使用就讀比例作為y軸
plt.figure(figsize=(12,6))#設定圖表大小
plt.bar(x,y,label="student%",width=0.4,align = "edge",color="#804040")#柱狀圖的元素
plt.title("2011-2014 NTU student")#圖表標題
plt.xticks(rotation=90)
plt.legend()
```
